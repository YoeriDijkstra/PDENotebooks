## In this notebook

In this notebook the heat equation

$$
\frac{\partial u}{\partial t}=a^2\frac{\partial ^2u}{\partial x^2}
$$

for $x\in (0,L),\ t>0$ is considered, where $a^2$ is is the thermal conductivity of a bar.

The initial and boundary conditions are

$$
\begin{align*}
u(x,0) &= f(x), \\
u(0,t) &= \phi(t), \\
u(L,t) &= \eta.
\end{align*}
$$

This notebook constructs a (truncated) Fourier series of the exact solution $u(x,t)$ of this problem using <code>sympy</code>, and then uses <code>numpy</code> and <code>pyplot</code> to display the obtained *exact* values.

The next cell imports a function that performs this.
